```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Öğrenci' : ['Ayşe','Fatma','Ali','Ömer','Buse','Evren','Zeynep'],
    'Matematik': [85,65,35,25,95,90,75],
    'Fizik' : [42,51,38,64,75,30,15],
    'Yaş' : [17,18,17,18,17,19,19],
    'Sınıf' :['11A','11B','11C','11B','11B','11A','11C']
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
# head() ilk 5 veriyi listeler
df.head(3)
```

```python
#tail() sondan veri listelemek için kullanuılır.
df.tail(2)
```

```python
df
```

```python
df.describe()
```

```python
# Sütun Seçme
df['Matematik']
```

```python
# birden fazla kolon seçme işlemi
# select "Öğrenci", "Matematik", "Fizik" from Öğrenciler;
df[['Öğrenci','Matematik','Fizik']]
```

```python
# Satır bazında filtreleme
# Matematik puanlarından 80 den fazla olan öğrenci bilgilerini listeleyiniz.
# SELECT * FROM ÖĞRENCİLER WHERE MATEMATİK > 80;

df[df['Matematik'] > 80 ]
```

```python
# Matematik dersinden 70 den fazla öğrenciler ve aynı zamanda Fizikten de 70 den fazla alan öğrencilerin listesi.
# and -> &
# or -> |
df[(df['Matematik']>70) & (df['Fizik']>70)]
```

```python
df[(df['Matematik']>70) | (df['Fizik']>70)]
```

```python
# 11A sınıfındaki öğrencilerin listesi
df[df['Sınıf']=='11A']
```

```python
# Kolon ekleme
df['Ortalama'] = (df['Matematik'] + df['Fizik'])/2
df
```

```python
df['Cinsiyet'] = ['K','K','E','E','K','E','K']
df
```

```python
# Ortalama sütununa göre artan bir şekilde sıralayınız.
df.sort_values('Ortalama')
```

```python
# Ortalama sütununa göre azalan bir şekilde sıralayınız.
df.sort_values('Ortalama',ascending=False)
```

```python
df.sort_values('Öğrenci',ascending=False)
# asc -> Artan
#desc -> Azalan
```

```python
# Gruplama
# mean() count() min() max() sum()
```

```python
# Sınıflara göre ortalama hesaplama
df.groupby('Sınıf')['Ortalama'].mean()
```

```python
# Sınıflara göre öğrenci sayısı
df.groupby('Sınıf')['Öğrenci'].count()
```

```python
# Cinsiyete göre kaç adet öğrenci vardır ?
df.groupby('Cinsiyet')['Öğrenci'].count()
```

```python
# Yaş alanaına göre ortalama not ortalamasını gösteriniz.
df.groupby('Yaş')['Ortalama'].mean()
```

```python
# Yaş alanaına göre  not ortalaması en büyük olan notları gösteriniz
df.groupby('Yaş')['Ortalama'].max()
```

```python
# En başarılı öğrenci
df.loc[df['Ortalama'].idxmax()]
```

```python
df['Ortalama'].idxmax()
```

```python
df['Ortalama'].idxmin()
```

```python
# En düşük matematik notu
df.loc[df['Matematik'].idxmin()]
```

```python
# sınıf bazında başarı sıralaması
df.groupby('Sınıf')['Ortalama'].mean().sort_values(ascending=False)
```
