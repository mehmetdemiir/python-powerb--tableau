```python
# saf python ile kodlayalım.
baslangic = time.time()
A = list(range(10000001))
B = list(range(10000001))
C = []
for i in range(len(B)):
  toplam = A[i] + B[i]
  C.append(toplam)
bitis = time.time()
sure = bitis- baslangic
print('Saf python ile çalışma süresi:',sure)
```
