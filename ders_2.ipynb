```python
# sum() : bir liste içerisinde ki tüm sayıları toplayan fonksiyondur.
liste1 = [10,2,3,4,1,5,6,21,14,14,25,36]
print(sum(liste1))
```

```python
min_deger = min(liste1)
print(min_deger)
```

```python
print(max(liste1))
```

```python
# aritmetik ortalamasını hesaplayınız.
toplam = sum(liste1)
eleman_sayisi = len(liste1)
ortalama = toplam / eleman_sayisi
print(ortalama)
```

```python
# Liste fonksiyonları :
#append(veri) : İlgili listeye sondan veri ekleme işlemi yapar.
```

```python
teams = ['Galatasaray','Fenerbahçe','Beşiktaş','Trabzonspor']
```

```python
print(teams)
```

```python
teams.append('Yeni Malatyaspor')
print(teams)
```

```python
# Kullanıcıdan kaç adet takım girilmesi isteniyorsa o bilgiyi alalım ve o takım kadar takım ekleyelim.
takim_sayisi = int(input('Kaç adet takım girmek istersiniz. '))
for i in range(takim_sayisi):
  takim_adi = input('Takım Adı : ')
  teams.append(takim_adi)
```

```python
print(teams)
```

```python
# remove(veri) : liste içerisinde ilgili veriyi silmek için kullanılır.
teams.remove('Karşıyaka')
print(teams)
```

```python
# sort artan veya a dan z ye sıralama işlemleri yapar.
teams.sort()
```

```python
print(teams)
```

```python
teams.sort(reverse=True)
print(teams)
```
